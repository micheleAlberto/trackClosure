```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import trackClosure.apps.BenchmarkApp as BE
be=BE.load_benchmark('data/bench0.be')
```

```python
cc_ids=be.CC.points.keys()
ConnectedComponents=pd.DataFrame.from_dict({
    'connected component':[cc_id for cc_id in cc_ids],
    'images': [ len(be.CC.points[cc_id].views) for cc_id in cc_ids],
    'keypoints': [ len(be.CC.points[cc_id].allViews()) for cc_id in cc_ids]
                        })
ConnectedComponents['keypointPairs']=ConnectedComponents.keypoints**2
```

```python
ConnectedComponents
```

```python
print '{} connected components observed over {} images group a total of {} keypoints whose {} keypoint pairs are to be verified'.format(
    len(ConnectedComponents),
    ConnectedComponents.images.max(),
    ConnectedComponents.keypoints.sum(),
    ConnectedComponents.keypointPairs.sum())
```

```python
cc_ids=be.oracle.points.keys()
oracles=pd.DataFrame.from_dict({
    'cc_id':[cc_id for cc_id in cc_ids],
    'images': [ len(be.oracle.points[cc_id].views) for cc_id in cc_ids],
    'keypoints': [ len(be.oracle.points[cc_id].allViews()) for cc_id in cc_ids]
                        })
oracles['positives']=oracles.keypoints**2
oracles
```

```python
print '{} oracles mark {} keypoints for a total of {} ground truth positives'.format(
    len(cc_ids),
    oracles.keypoints.sum(),
    oracles.positives.sum())
```
